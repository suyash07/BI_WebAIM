#### Creating panda series for each wave category and technology


```python
million_metrics = million.iloc[:,0:8]
wcag_fail = million.iloc[:,9:11]
#wave_contrast = million['contrast']
wave_errors = million.iloc[:,11:34]
wave_alerts = million.iloc[:,34:68]
wave_features = million.iloc[:,68:81]
wave_totalaria = million.iloc[:,81]
wave_html5_aria = million.iloc[:,82:92]
wave_structural = million.iloc[:,92:108]
technologies = million.loc[:,'Ads - AdSense':]
```

### How do technologies align with certain error types?

```python
technologies_by_errors = {}
errors_tech = pd.merge(wave_errors,technologies, how='right',left_index = True, right_index = True)

for i in range(len(technologies.columns)):
    tech_type = errors_tech[technologies.columns[i]] == 1
    tech_with_errors = errors_tech[tech_type]
    tech_with_errors.loc['Total'] = tech_with_errors.sum()
    top_errors = tech_with_errors.loc['Total']
    ratio_errors = tech_with_errors.loc['Total']/top_errors.loc[technologies.columns[i]]
    technologies_by_errors[technologies.columns[i]] = ratio_errors

error_analysis = pd.DataFrame(technologies_by_errors)
```

```python
errors_tech = {}
errors_tech['Technologies'] = []
errors_tech['Error Type'] = []
errors_tech['Value'] = []


for i in range(len(technologies.columns)):
    for j in range(len(wave_errors.columns)):
        errors_tech['Technologies'].append(technologies.columns[i])
        errors_tech['Error Type'].append(wave_errors.columns[j])
        errors_tech['Value'].append(error_analysis.loc[wave_errors.columns[j],technologies.columns[i]])
        
errors_and_tech_analysis = pd.DataFrame(errors_tech)
errors_and_tech_analysis.head()
```

```python
errors_and_tech_analysis.to_csv('error and tech analysis.csv')
```

### How do technologies align with certain alert types?

```python
technologies_by_alerts = {}
alerts_tech = pd.merge(wave_alerts,technologies, how='right',left_index = True, right_index = True)

for i in range(len(technologies.columns)):
    tech_type = alerts_tech[technologies.columns[i]] == 1
    tech_with_alerts = alerts_tech[tech_type]
    tech_with_alerts.loc['Total'] = tech_with_alerts.sum()
    top_alerts = tech_with_alerts.loc['Total']
    ratio_errors = tech_with_alerts.loc['Total']/top_alerts.loc[technologies.columns[i]]
    technologies_by_alerts[technologies.columns[i]] = ratio_errors

alerts_analysis = pd.DataFrame(technologies_by_alerts)
```

```python
alerts_tech = {}
alerts_tech['Technologies'] = []
alerts_tech['Alert Type'] = []
alerts_tech['Value'] = []

for i in range(len(technologies.columns)):
    for j in range(len(wave_alerts.columns)):
        alerts_tech['Technologies'].append(technologies.columns[i])
        alerts_tech['Alert Type'].append(wave_alerts.columns[j])
        alerts_tech['Value'].append(alerts_analysis.loc[wave_alerts.columns[j],technologies.columns[i]])
        

alerts_and_tech_analysis = pd.DataFrame(alerts_tech)
alerts_and_tech_analysis.head()
```

```python
technologies_by_features = {}
features_tech = pd.merge(wave_features,technologies, how='right',left_index = True, right_index = True)

for i in range(len(technologies.columns)):
    tech_type = features_tech[technologies.columns[i]] == 1
    tech_with_features = features_tech[tech_type]
    tech_with_features.loc['Total'] = tech_with_features.sum()
    top_features = tech_with_features.loc['Total']
    ratio_errors = tech_with_features.loc['Total']/top_features.loc[technologies.columns[i]]
    technologies_by_features[technologies.columns[i]] = ratio_errors

features_analysis = pd.DataFrame(technologies_by_features)
```

```python
features_tech = {}
features_tech['Technologies'] = []
features_tech['Feature Type'] = []
features_tech['Value'] = []

for i in range(len(technologies.columns)):
    for j in range(len(wave_features.columns)):
        features_tech['Technologies'].append(technologies.columns[i])
        features_tech['Feature Type'].append(wave_features.columns[j])
        features_tech['Value'].append(features_analysis.loc[wave_features.columns[j],technologies.columns[i]])
        

features_and_tech_analysis = pd.DataFrame(features_tech)
features_and_tech_analysis.head()
```
